```python
import pandas as pd
import numpy as np
```

```python
# حتى تكون النتائج نفسها كل مرة
np.random.seed(42)

# عدد الشركات
num_companies = 150
```

```python
companies = [f"Company_{i}" for i in range(1, num_companies + 1)]

sectors = np.random.choice(
    [
        "Banking",
        "Retail",
        "Energy",
        "Telecom",
        "Manufacturing",
        "Healthcare",
        "Technology"
    ],
    num_companies
)
```

```python
total_assets = np.random.randint(5_000_000, 100_000_000, num_companies)

total_liabilities = (
    total_assets *
    np.random.uniform(0.2, 0.9, num_companies)
).astype(int)

revenue = np.random.randint(2_000_000, 80_000_000, num_companies)

net_income = (
    revenue *
    np.random.uniform(0.02, 0.25, num_companies)
).astype(int)

cash_flow = (
    net_income *
    np.random.uniform(0.8, 1.3, num_companies)
).astype(int)

current_assets = (
    total_assets *
    np.random.uniform(0.2, 0.6, num_companies)
).astype(int)

current_liabilities = (
    total_liabilities *
    np.random.uniform(0.3, 0.8, num_companies)
).astype(int)
```

```python
risk = np.random.choice(
    ["Low", "Medium", "High"],
    num_companies,
    p=[0.4, 0.35, 0.25]
)
```

```python
df = pd.DataFrame({
    "Company": companies,
    "Sector": sectors,
    "Total_Assets": total_assets,
    "Total_Liabilities": total_liabilities,
    "Revenue": revenue,
    "Net_Income": net_income,
    "Cash_Flow": cash_flow,
    "Current_Assets": current_assets,
    "Current_Liabilities": current_liabilities,
    "Risk": risk
})
```
